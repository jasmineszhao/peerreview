```python
from pathlib import Path
import pandas as pd

kff_path = Path("/home/jszhao/econ470/a0/work/peerreview/data/input/raw_data.csv")

kff = pd.read_csv(kff_path, skiprows=2)
kff = kff[kff["Location"] != "United States"].copy()

kff.head()
kff.columns.tolist()
```

```python
import numpy as np

kff = kff.rename(columns={"Location": "state_name"})

kff["expansion_date"] = pd.to_datetime(
    kff["Expansion Implementation Date"],
    errors="coerce"
)

kff["expansion_year"] = kff["expansion_date"].dt.year
kff["expanded"] = (kff["Status of Medicaid Expansion Decision"] == "Adopted").astype(int)
kff["never_expanded"] = (kff["Status of Medicaid Expansion Decision"] == "Not Adopted").astype(int)
kff["expanded_2014"] = (kff["expansion_year"] == 2014).astype(int)

kff = kff[[
    "state_name",
    "Status of Medicaid Expansion Decision",
    "expansion_date",
    "expansion_year",
    "expanded",
    "never_expanded",
    "expanded_2014"
]].copy()

kff.head(10)
```

```python
state_crosswalk = pd.DataFrame({
    "state": [
        "AL","AK","AZ","AR","CA","CO","CT","DE","DC","FL",
        "GA","HI","ID","IL","IN","IA","KS","KY","LA","ME",
        "MD","MA","MI","MN","MS","MO","MT","NE","NV","NH",
        "NJ","NM","NY","NC","ND","OH","OK","OR","PA","RI",
        "SC","SD","TN","TX","UT","VT","VA","WA","WV","WI","WY"
    ],
    "state_name": [
        "Alabama","Alaska","Arizona","Arkansas","California","Colorado","Connecticut","Delaware","District of Columbia","Florida",
        "Georgia","Hawaii","Idaho","Illinois","Indiana","Iowa","Kansas","Kentucky","Louisiana","Maine",
        "Maryland","Massachusetts","Michigan","Minnesota","Mississippi","Missouri","Montana","Nebraska","Nevada","New Hampshire",
        "New Jersey","New Mexico","New York","North Carolina","North Dakota","Ohio","Oklahoma","Oregon","Pennsylvania","Rhode Island",
        "South Carolina","South Dakota","Tennessee","Texas","Utah","Vermont","Virginia","Washington","West Virginia","Wisconsin","Wyoming"
    ]
})
```

```python
hcris["state"] = hcris["state"].astype(str).str.strip().str.upper()

hcris = hcris.merge(state_crosswalk, on="state", how="left")
analytic = hcris.merge(kff, on="state_name", how="left")

analytic.shape
```

```python
analytic["uncomp_care"] = analytic["tot_uncomp_care_charges"] - analytic["tot_uncomp_care_partial_pmts"]
analytic["uncomp_care_mil"] = analytic["uncomp_care"] / 1_000_000

analytic["treat"] = analytic["expanded_2014"]
analytic["post"] = (analytic["year"] >= 2014).astype(int)

analytic["sample_2014_vs_never"] = (
    (analytic["expanded_2014"] == 1) | (analytic["never_expanded"] == 1)
).astype(int)
```
